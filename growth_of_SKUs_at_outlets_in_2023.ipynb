```python
# import
import pandas as pd
import duckdb
```

```python
# read outlet data 2023
ol_df1 = pd.read_excel(open("C:/Users/Shithi.Maitra/Downloads/WS SKU Wise Outlet wise data.xlsx", "rb"), sheet_name="2023", header=1, index_col=None)
ol_df1.columns = ['outlet_code', 'basepack', 'jan_23', 'feb_23', 'mar_23', 'apr_23', 'may_23', 'jun_23', 'jul_23', 'aug_23', 'sep_23', 'total_23']

# refine outlet codes

# seq
ol_df1 = duckdb.query('''select *, row_number() over() seq from ol_df1''').df()

# min
qry = '''
select outlet_code outlet_code_rf, min(seq) min_seq
from ol_df1
where outlet_code is not null
group by 1
order by 2 asc
'''
df = duckdb.query(qry).df()

# max
min_seq = df['min_seq'].tolist()
max_seq = []
entries = len(min_seq)
for i in range(1, entries): max_seq.append(min_seq[i]-1)
max_seq.append(ol_df1.shape[0])
df['max_seq'] = max_seq

# join
qry = '''
select outlet_code_rf outlet_code, basepack, jan_23, feb_23, mar_23, apr_23, may_23, jun_23, jul_23, aug_23, sep_23, total_23
from 
    ol_df1 tbl1 
    inner join 
    df tbl2 on(seq>=min_seq and seq<=max_seq)
'''
ol_df1 = duckdb.query(qry).df()

# handle nulls
ol_df1 = ol_df1.fillna(0)

# see
display(ol_df1)
```

```python
# read outlet data 2022
ol_df2 = pd.read_excel(open("C:/Users/Shithi.Maitra/Downloads/WS SKU Wise Outlet wise data.xlsx", "rb"), sheet_name="2022", header=1, index_col=None)
ol_df2.columns = ['outlet_code', 'basepack', 'jan_22', 'feb_22', 'mar_22', 'apr_22', 'may_22', 'jun_22', 'jul_22', 'aug_22', 'sep_22', 'oct_22', 'nov_22', 'dec_22', 'total']

# partial
ol_df2 = ol_df2[['outlet_code', 'basepack', 'jan_22', 'feb_22', 'mar_22', 'apr_22', 'may_22', 'jun_22', 'jul_22', 'aug_22', 'sep_22', 'dec_22']]

# refine outlet codes

# seq
ol_df2 = duckdb.query('''select *, row_number() over() seq from ol_df2''').df()

# min
qry = '''
select outlet_code outlet_code_rf, min(seq) min_seq
from ol_df2
where outlet_code is not null
group by 1
order by 2 asc
'''
df = duckdb.query(qry).df()

# max
min_seq = df['min_seq'].tolist()
max_seq = []
entries = len(min_seq)
for i in range(1, entries): max_seq.append(min_seq[i]-1)
max_seq.append(ol_df2.shape[0])
df['max_seq'] = max_seq

# join
qry = '''
select outlet_code_rf outlet_code, basepack, jan_22, feb_22, mar_22, apr_22, may_22, jun_22, jul_22, aug_22, sep_22, dec_22
from 
    ol_df2 tbl1 
    inner join 
    df tbl2 on(seq>=min_seq and seq<=max_seq)
'''
ol_df2 = duckdb.query(qry).df()

# handle nulls
ol_df2 = ol_df2.fillna(0)

# add
ol_df2['total_22'] = ol_df2['jan_22'] + ol_df2['feb_22'] + ol_df2['mar_22'] + ol_df2['apr_22'] + ol_df2['may_22'] + ol_df2['jun_22'] + ol_df2['jul_22'] + ol_df2['aug_22'] + ol_df2['sep_22'] 

# see
display(ol_df2)
```

```python
# combine
qry = '''
select * 
from 
    ol_df1 tbl1
    full join 
    ol_df2 tbl2 using(outlet_code, basepack)
'''
ol_df = duckdb.query(qry).df().fillna(0)
display(ol_df)
```

```python
# union
qry = '''
select outlet_code, basepack, dec_22, jan_23, feb_23, mar_23, apr_23, may_23, jun_23, jul_23, aug_23, sep_23, total_23, total_22, (total_23-total_22)*1.00/total_22 growth, 1 typ
from ol_df
union all
select *, jan_23+feb_23+mar_23+apr_23+may_23+jun_23+jul_23+aug_23+sep_23 total_23, 0 total_22, 0 growth, 2 typ
from
    (select 
        outlet_code, 
        basepack,
        0 dec_22, 
        jan_23-dec_22 jan_23, 
        feb_23-jan_23 feb_23, 
        mar_23-feb_23 mar_23, 
        apr_23-mar_23 apr_23, 
        may_23-apr_23 may_23, 
        jun_23-may_23 jun_23, 
        jul_23-jun_23 jul_23, 
        aug_23-jul_23 aug_23, 
        sep_23-aug_23 sep_23
    from ol_df
    ) tbl1
'''
tbl = duckdb.query(qry).df()

# BC
qry = '''
select outlet_code, sum(total_23)*1.00/(select sum(total_23) from tbl where typ=1) outlet_bc
from tbl
where typ=1
group by 1
'''
ol_bc_df = duckdb.query(qry).df()
qry = '''
select basepack, sum(total_23)*1.00/(select sum(total_23) from tbl where typ=1) basepack_bc
from tbl
where typ=1
group by 1
'''
bp_bc_df = duckdb.query(qry).df()

# growth
qry = '''
select outlet_code, sum(total_23-total_22)*1.00/sum(total_22) outlet_growth
from tbl
where typ=1
group by 1
'''
ol_growth_df = duckdb.query(qry).df()
qry = '''
select basepack, sum(total_23-total_22)*1.00/sum(total_22) basepack_growth
from tbl
where typ=1
group by 1
'''
bp_growth_df = duckdb.query(qry).df()
```

```python
# OP
qry = '''
select outlet_code, basepack, dec_22, jan_23, feb_23, mar_23, apr_23, may_23, jun_23, jul_23, aug_23, sep_23, total_23, total_22, growth, outlet_bc, basepack_bc, typ
from 
    tbl tbl1 
    inner join 
    ol_bc_df using(outlet_code)
    inner join 
    bp_bc_df using(basepack)
order by outlet_bc desc, outlet_code asc, basepack_bc desc, basepack asc, typ asc
'''
res_df1 = duckdb.query(qry).df()
display(res_df1)
```

```python
# summary

# outlet
qry = '''
select outlet_code, total, total_diff, outlet_bc, outlet_growth
from 
    (select outlet_code, sum(total_23) total, max(outlet_bc) outlet_bc
    from res_df1
    where typ=1
    group by 1
    ) tbl1 
    inner join 
    (select outlet_code, sum(total_23) total_diff
    from res_df1
    where typ=2
    group by 1
    ) tbl2 using(outlet_code)
    inner join 
    ol_growth_df tbl3 using(outlet_code)
order by outlet_bc desc
'''
res_df2 = duckdb.query(qry).df()
display(res_df2)

# basepack
qry = '''
select basepack, total, total_diff, basepack_bc, basepack_growth
from 
    (select basepack, sum(total_23) total, max(basepack_bc) basepack_bc
    from res_df1
    where typ=1
    group by 1
    ) tbl1 
    inner join 
    (select basepack, sum(total_23) total_diff
    from res_df1
    where typ=2
    group by 1
    ) tbl2 using(basepack)
    inner join 
    bp_growth_df tbl3 using(basepack)
order by basepack_bc desc
'''
res_df3 = duckdb.query(qry).df()
display(res_df3)
```

```python
# store
with pd.ExcelWriter("C:/Users/Shithi.Maitra/Downloads/outlet_basepack_delta_file.xlsx") as writer:
    res_df = duckdb.query('''select * from res_df1 limit 1000000''').df()
    res_df.to_excel(writer, sheet_name="Differences-BC", index=False)
    res_df2.to_excel(writer, sheet_name="Outlet Summary", index=False)
    res_df3.to_excel(writer, sheet_name="Basepack Summary", index=False)
```
